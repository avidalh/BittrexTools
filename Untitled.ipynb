```python
import pandas as pd
import matplotlib.pyplot as plt
import json

%config InlineBackend.figure_format='retina'
%matplotlib inline

# url to pick historical data:
# https://bittrex.com/Api/v2.0/pub/market/GetTicks?marketName=BTC-ZEC&tickInterval=oneMin&_=1499127220008

filename = './USDT-ZEC_thirtyMin.json'

# loads json data into dataframe, set datetime as table index
with open(filename) as data_file:
    data = json.load(data_file)
data = data['result']
df = pd.DataFrame.from_dict(data)
df = df.set_index(pd.DatetimeIndex(df['T']) )

df.info()
```

```python
def STOK(close, low, high, n): 
 STOK = ((close - pd.rolling_min(low, n)) / (pd.rolling_max(high, n) - pd.rolling_min(low, n))) * 100
 return STOK

def STOD(close, low, high, n):
 STOK = ((close - pd.rolling_min(low, n)) / (pd.rolling_max(high, n) - pd.rolling_min(low, n))) * 100
 STOD = pd.rolling_mean(STOK, 3)
 return STOD

df['%K'] = STOK(df['C'], df['L'], df['H'], 14)
df['%D'] = STOD(df['C'], df['L'], df['H'], 14)
df.tail()

df[-100:].plot(y=['C'], figsize = (10, 5), grid=True)
df[-100:].plot(y=['%K', '%D'], figsize = (10, 5), grid=True)
```

```python
import pandas as pd
import numpy as np
from pandas_datareader import data as web
import matplotlib.pyplot as plt
%matplotlib inline

def get_stock(stock,start,end):
 return web.DataReader(stock,'google',start,end)['Close']

def get_high(stock,start,end):
 return web.DataReader(stock,'google',start,end)['High']

def get_low(stock,start,end):
 return web.DataReader(stock,'google',start,end)['Low']

def STOK(close, low, high, n): 
 STOK = ((close - pd.rolling_min(low, n)) / (pd.rolling_max(high, n) - pd.rolling_min(low, n))) * 100
 return STOK

def STOD(close, low, high, n):
 STOK = ((close - pd.rolling_min(low, n)) / (pd.rolling_max(high, n) - pd.rolling_min(low, n))) * 100
 STOD = pd.rolling_mean(STOK, 3)
 return STOD

df = pd.DataFrame(get_stock('FB', '1/1/2016', '12/31/2016'))
df['High'] = get_high('FB', '1/1/2016', '12/31/2016')
df['Low'] = get_low('FB', '1/1/2016', '12/31/2016')
df['%K'] = STOK(df['Close'], df['Low'], df['High'], 14)
df['%D'] = STOD(df['Close'], df['Low'], df['High'], 14)
df.tail()
```

```python
df.plot(y=['Close'], figsize = (20, 5))
df.plot(y=['%K', '%D'], figsize = (20, 5))
```
